  Input: `lst = [5, 0, 1]`

#### Starter Code for Question 3


```python
from typing import List, Union

def missing_num(nums: List) -> Union[List[int], int]:
    if not nums:
        return -1
    max_val = max(nums)
    full_set = set(range(max_val + 1))
    missing = sorted(full_set - set(nums))
    return missing if missing else -1
```

```python
lst = [6,8,2,5,7,0,1]
missing_num(lst)
```

```python
# Your answer here



# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, val = 0, left = None, right = None):
        self.val = val
        self.left = left
        self.right = right
        
def is_duplicate(root: TreeNode) -> int:
  if root is None:
    return False
  if root.left is not None and root.left.val == root.val:
    return True
  if root.right is not None and root.right.val == root.val:
    return True
  return is_duplicate(root.left) or is_duplicate(root.right)
  
  
  

```


-   Explain why their solution works in your own words.


The code looks at each pair of parent and child in the tree one at a time. As soon as it finds a child with the same number as its parent, it detects a duplicate and stops. If a node has no children, it moves on to the next. By checking every parent–child link once and stopping at the first match, it guarantees you get the duplicate that’s closest to the top of the tree.


-   Explain the problem’s time and space complexity in your own words.


The code checks every node just one time. That means if we double the nodes, it will take about twice as long. This means the time taken to process is directly proprotional to the number of nodes, indicating that it is O(n) time. As it goes, it keeps track of the route from the top down, so the deeper the tree, the more memory it uses. When the tree is so deep that it is a straight line of nodes, the route the algorithm travels is as long as the tree (in this case it is O(n) space). If it is a shorter tree, it will be closer to O(log n) space since the depth will be on par with the height of the tree.
